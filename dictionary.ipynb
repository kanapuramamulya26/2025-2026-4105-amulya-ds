```python
# DICTIONARY OPERATIONS

# Creating a dictionary
d = {
    "name": "K.Amulya",
    "age": 20,
    "course": "BTech"
}
print("Original dictionary:", d)

# Accessing values
print("Name:", d["name"])
print("Age:", d.get("age"))

# Adding a new key-value pair
d["college"] = "XYZ University"
print("After adding:", d)

# Updating value
d["age"] = 21
print("After updating:", d)

# Removing elements
d.pop("course")
print("After pop:", d)

removed_item = d.popitem()
print("After popitem:", d)
print("Removed item:", removed_item)

# Keys, values, items
print("Keys:", d.keys())
print("Values:", d.values())
print("Items:", d.items())

# Membership
print("Is 'name' a key?", "name" in d)

# Length
print("Length of dictionary:", len(d))

# Looping through dictionary
print("Looping through keys:")
for key in d:
    print(key, ":", d[key])

print("Looping through key-value pairs:")
for k, v in d.items():
    print(k, "->", v)

# Copying dictionary
d_copy = d.copy()
print("Copied dictionary:", d_copy)

# Updating using update()
d.update({"city": "Hyderabad", "age": 22})
print("After update():", d)

# Nested dictionary
student = {
    "id": 1,
    "info": {
        "name": "Amit",
        "marks": 90
    }
}
print("Nested dictionary value:", student["info"]["marks"])

# Clearing dictionary
d.clear()
print("After clear:", d)
```
