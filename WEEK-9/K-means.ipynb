34) Write a code to implement K-means clustering (unsupervised learning) algorithm

```python
import numpy as np
A=np.array([[2,3],[3,4],[4,2]])
B=np.array([[8,7],[9,6],[10,8]])
P=np.array([5,5])
cA=A.mean(axis=0)
cB=B.mean(axis=0)
dA=np.linalg.norm(P-cA)
dB=np.linalg.norm(P-cB)
if dA<dB:
  print("Point belongs to cluster A")
else:
    print("point belongs to cluster B")
```
