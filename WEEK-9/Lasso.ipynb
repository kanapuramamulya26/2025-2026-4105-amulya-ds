```python
import numpy as np
from sklearn.linear_model import Lasso
X=np.array([[1,5,2],[2,6,4],[4,3,6],[7,6,2]])
Y=np.array([5,10,15,20])
model=Lasso(alpha=0.1)
model.fit(X,Y)
print("coefficient:",model.coef_)
```
