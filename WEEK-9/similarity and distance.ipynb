33) Implement the similarities and distances for any dataset

```python
import numpy as np
from scipy.spatial import distance
from scipy.stats import pearsonr
p1=np.array([2,3])
p2=np.array([5,8])
euc=distance.euclidean(p1,p2)
mh=distance.cityblock(p1,p2)
cb=distance.chebyshev(p1,p2)
hm=distance.hamming(p1,p2)
cd=distance.cosine(p1,p2)
cs=1-cd
set1,set2=set(p1),set(p2)
js=len(set1.intersection(set2))/len(set1.union(set2))
r,_=pearsonr(p1,p2)
print("Euclidean Distance:",euc)
print("Manhattan Distance:",mh)
print("Chebyshev Distance:",cb)
print("Hamming Distance:",hm)
print("Cosine similarity:",cs)
print("Pearsonr correlation:",r)
print("Jaccard similarity:",js)
```
