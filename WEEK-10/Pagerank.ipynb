37) Write the code to implement the page rank algorithm

```python
import networkx as nx
G=nx.DiGraph()
G.add_edges_from ([
('A', 'B'),
('A', 'C'),
('B', 'C'),
('D', 'C'),
('E', 'B'),
('E','C')
])
pagerank=nx.pagerank (G, alpha=0.85)
print ("PageRank Scores:")
for page, rank in pagerank.items():
 print(f" {page} : {rank : .4f} ")
```
