31. Implement random forest for given dataset


```python
from sklearn.ensemble import RandomForestClassifier
x=[[1,1],[1,0],[0,1],[0,0]]
y=[0,0,1,1]
model=RandomForestClassifier()
model.fit(x,y)
print(" Model prediction:",model.predict([[1,0]]))
```
