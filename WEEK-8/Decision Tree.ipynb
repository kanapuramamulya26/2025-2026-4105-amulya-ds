30. Implement the decision tree for the given data set

```python
from sklearn.tree import DecisionTreeClassifier
x=[[1,1],[1,0],[0,1],[0,0]]
y=[0,1,1,0]
model=DecisionTreeClassifier()
model.fit(x,y)
print("Model prediction:",model.predict([[1,0]]))
```
