27) Implement the confusion matrix for any of the example dataset

```python
import pandas as pd
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
data={
    'marks':[1,0,1,0,1,0,1,0],
    'grade':[1,1,1,0,0,0,1,0]
}
df=pd.DataFrame(data)
y_true=df['marks']
y_pred=df['grade']
cm=confusion_matrix(y_true,y_pred)
print("\n Confusion matrix:",cm)
```
