28. Write the code to implement support vector machine(SVM)


```python
from sklearn import svm
x=[[2,2],[4,4],[2,0],[0,2]]
y=[1,1,-1,-1]
model=svm.SVC(kernel='linear')
model.fit(x,y)
print(model.predict([[2,0]]))
```
