24.Write a code to implement linear regression with an example URL

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
url="https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv"
data=pd.read_csv(url)
print("Dataset Preview:")
print(data.head())
x=data[["Experience Years"]]
y=data["Salary"]
model=LinearRegression()
model.fit(x, y)
print("\nModel Parameters:")
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
years=[[5]]
predicted_salary=model.predict(years)
print("\nPredicted salary for 5 years experience:", predicted_salary[0])
y_pred=model.predict(x)
mse=mean_squared_error(y, y_pred)
print("\nMean Squared Error:", mse)
plt.scatter(x, y, color='blue')
plt.plot(x, y_pred, color='red')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression Salary Prediction")
plt.show()
```
